### ðŸŽ¯ Objective:
To analyze global trends in AI adoption, economic impact, industry distribution, consumer trust, and policy influence using a synthesized dataset across multiple countries and industries.


```python
# Generating more lines of data based on my sample data 

import pandas as pd
import numpy as np

# Load original CSV
df = pd.read_csv("E:/python_cybrom/Global_AI_Content_Impact_Dataset.csv")

# Unique values to sample from
countries = df['Country'].unique()
industries = df['Industry'].unique()
tools = df['Top AI Tools Used'].unique()
regulations = df['Regulation Status'].unique()
years = np.arange(2020, 2026)

# Function to generate synthetic row
def generate_row():
    return {
        'Country': np.random.choice(countries),
        'Year': np.random.choice(years),
        'Industry': np.random.choice(industries),
        'AI Adoption Rate (%)': np.round(np.random.uniform(20, 95), 2),
        'AI-Generated Content Volume (TBs per year)': np.round(np.random.uniform(10, 120), 2),
        'Job Loss Due to AI (%)': np.round(np.random.uniform(5, 50), 2),
        'Revenue Increase Due to AI (%)': np.round(np.random.uniform(0, 80), 2),
        'Human-AI Collaboration Rate (%)': np.round(np.random.uniform(10, 90), 2),
        'Top AI Tools Used': np.random.choice(tools),
        'Regulation Status': np.random.choice(regulations),
        'Consumer Trust in AI (%)': np.round(np.random.uniform(25, 70), 2),
        'Market Share of AI Companies (%)': np.round(np.random.uniform(1, 40), 2)
    }

# Generate 800 synthetic rows
new_data = [generate_row() for _ in range(800)]
df_new = pd.DataFrame(new_data)

# Combine with original
df_combined = pd.concat([df, df_new], ignore_index=True)

# Save to new CSV
df_combined.to_csv("Expanded_Global_AI_Content_Dataset.csv", index=False)
```

<h3 style = "color:yellow">--> AI Adoption Trends</h3>

```python
sns.barplot(data=df, x='Year', y='AI Adoption Rate (%)')
plt.title("AI Adoption Rate Over Years by Country")
plt.show()
```

```python
sns.set(style='ticks')
g = sns.FacetGrid(df, col="Industry", col_wrap=3, height=4)
g.map(plt.hist, "AI Adoption Rate (%)", bins=10, color='skyblue')
g.fig.subplots_adjust(top=0.9)
g.fig.suptitle("Distribution of AI Adoption Rate (%) Across Industries")
plt.show()
```

```python
g = sns.FacetGrid(df, col="Regulation Status", height=4)
g.map(plt.hist, "Job Loss Due to AI (%)", bins=10, color='salmon')
g.fig.subplots_adjust(top=0.8)
g.fig.suptitle("Job Loss Distribution by Regulation Type")
plt.show()
```

<h3 style = "color:yellow">--> Top 10 Countries by Average AI Adoption</h3>

Helps identify global leaders in AI adoption.

```python
# Prepare the data
top_countries = df.groupby("Country")["AI Adoption Rate (%)"].mean().sort_values(ascending=False).head(10).reset_index()
top_countries['Color'] = top_countries['Country']  # for hue mapping

# Plot with hue + legend=False
plt.figure(figsize=(10,6))
sns.barplot(
    data=top_countries,
    x="AI Adoption Rate (%)",
    y="Country",
    hue="Color",
    dodge=False,
    palette="viridis",
    legend=False
)
plt.title("Top 10 Countries by Average AI Adoption Rate (%)")
plt.xlabel("Average AI Adoption Rate (%)")
plt.ylabel("Country")
plt.tight_layout()

plt.show()
```

<h3 style = "color:yellow">--> Industry-Wise Impact of AI on Jobs</h3>

Which industries are hit hardest in terms of job losses?

```python
plt.figure(figsize=(12,6))
sns.boxplot(data=df, x="Industry", y="Job Loss Due to AI (%)", palette="rocket")
plt.xticks(rotation=45)
plt.title("Distribution of Job Loss (%) by Industry")
plt.tight_layout()
plt.show()
```

```python
industry_avg = df.groupby("Industry")["Job Loss Due to AI (%)"].mean().sort_values()

plt.figure(figsize=(12,6))
sns.barplot(x=industry_avg.values, y=industry_avg.index, palette="rocket")
plt.title("Average Job Loss (%) by Industry")
plt.xlabel("Job Loss (%)")
plt.tight_layout()
plt.show()
```

<h3 style = "color:yellow">--> AI-Generated Content Volume Over Time (by Industry)</h3>

Shows the rise of AI content creation across sectors.

```python
plt.figure(figsize=(14,7))
sns.lineplot(data=df, x="Year", y="AI-Generated Content Volume (TBs per year)", hue="Industry", estimator='mean')
plt.title("AI-Generated Content Growth by Industry")
plt.ylabel("Content Volume (TBs/year)")
plt.tight_layout()
plt.show()
```

<h3 style = "color:yellow">--> Correlation Between Revenue Growth and AI Adoption</h3>

Reveals whether AI adoption is driving business performance.

```python
plt.figure(figsize=(8,6))
sns.regplot(data=df, x="AI Adoption Rate (%)", y="Revenue Increase Due to AI (%)", scatter_kws={"alpha":0.5}, line_kws={"color":"red"})
plt.title("Does AI Adoption Drive Revenue Growth?")
plt.tight_layout()
plt.show()
```

<h3 style = "color:yellow">--> Most Common AI Tools in Use</h3>

Visualizes popularity across the dataset.

```python
tool_counts = df['Top AI Tools Used'].value_counts()

plt.figure(figsize=(8,5))
sns.barplot(y=tool_counts.values, x=tool_counts.index, palette="magma")
plt.title("Most Popular AI Tools Used")
plt.xlabel("Count")
plt.ylabel("Tool")
plt.tight_layout()
plt.show()
```

<h3 style = "color:yellow">-->  Heatmap of Metric Correlations</h3>

Fast view of relationships between all numerical KPIs.

```python
plt.figure(figsize=(12,8))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='YlGnBu', linewidths=0.5)
plt.title("Correlation Matrix of AI Impact Metrics")
plt.tight_layout()
plt.show()
```

<h3 style = "color:yellow">--> Industry Representation in Dataset</h3>


```python
industry_counts = df['Industry'].value_counts()

plt.figure(figsize=(7,7))
plt.pie(industry_counts, labels=industry_counts.index, autopct='%1.1f%%', startangle=150, colors=sns.color_palette("Set2"))
plt.title("Industry Representation in Dataset")
plt.axis('equal')
plt.show()
```

<h3 style = "color:yellow">--> Overall AI Adoption Rate (Globally)</h3>


```python
# Group by year and calculate average AI adoption globally
yearly_trend = df.groupby("Year")["AI Adoption Rate (%)"].mean().reset_index()

# Plot the trend
plt.figure(figsize=(10,6))
sns.lineplot(data=yearly_trend, x="Year", y="AI Adoption Rate (%)", marker="o", linewidth=2.5, color="teal")

plt.title("Global Trend of AI Adoption Over the Years")
plt.ylabel("Average AI Adoption Rate (%)")
plt.xlabel("Year")
plt.grid(True)
plt.tight_layout()
plt.show()
```
